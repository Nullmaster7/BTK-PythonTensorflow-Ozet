```python
numpyDizisi1 = np.linspace(0,100,20)
```

```python
numpyDizisi2 = numpyDizisi1 ** 4
```

```python
plt.plot(numpyDizisi1, numpyDizisi2, "g*-") # g-- kesik çizgi, g*- yıldızlarla gösteriyor grafigi örneğin.
```

```python
plt.subplot(1,2,1) # shift tab description oku
plt.plot(numpyDizisi1, numpyDizisi2, "r*-")

plt.subplot(1,2,2) # 1 sıra olacak, 2 kolon olacak, şu an 2. grafigi çiziyorum
plt.plot(numpyDizisi2, numpyDizisi1, "g--")
```

```python
# Artık matplotlib kütüphanesinden olusturdugumuz nesne ile islem yapalım daha hakim olmak icin.
benimFigure = plt.figure() # boş bir canvas gibi düşün

figureAxes = benimFigure.add_axes([0.2, 0.2, 0.9, 0.9])
figureAxes.plot(numpyDizisi1, numpyDizisi2, "g")
figureAxes.set_xlabel("X ekseni veri ismi")
figureAxes.set_ylabel("Y ekseni veri ismi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7]) # deneyerek görebilirsin nereye nasıl etki ediyor.
eksen2 = figure2.add_axes([0.2,0.5,0.3,0.2])

eksen1.plot(numpyDizisi1, numpyDizisi2, "g")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Ana grafik başlığı")

eksen2.plot(numpyDizisi2, numpyDizisi1, "b")
eksen2.set_xlabel("X ekseni")
eksen2.set_ylabel("Y ekseni")
eksen2.set_title("Küçük Grafik Başlığı")
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows = 1, ncols=2)
type(benimEksenler) # type numpy array oluyor o yüzden döngüye aldım.

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1, numpyDizisi2, "g")
    eksen.set_xlabel("X ekseni")
    
plt.tight_layout() # sıgdırmaya yarıyor, iki grtafik birbirine girmesin diye
```

```python
yenifigure = plt.figure(figsize = (4,4), dpi = 100) # parametre verebiliriz. (figsize), (dpi) = dot per pixel

yeniEksen = yenifigure.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1, numpyDizisi2 ** 2, label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1, numpyDizisi2 ** 3, label = "numpyDizisi **3")
yeniEksen.legend(loc = 1) # loc değiştirebiliriz deneyerek ya da default bırakabiliriz.
```

## Gorsellik

```python
numpyDizi3 = np.linspace(0,10,20)
```

```python
numpyDizi4 = numpyDizi3 ** 2
```

```python
(figure, eksen) = plt.subplots()
eksen.plot(numpyDizi3, numpyDizi4, "g", color = "#3A95A8", alpha = 0.3) # renkler için internetten rgb hex code araştır.
eksen.plot(numpyDizi4, numpyDizi3, "b") # alpha saydamlık oranıdır.
```

```python
(yenifigure, yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizi3, numpyDizi3 + 2, color = "blue", linewidth = 5.0)
yeniEksen.plot(numpyDizi3, numpyDizi3 + 3, color = "yellow",linewidth = 2.5)


yeniEksen.plot(numpyDizi3, numpyDizi3 + 4, color = "red", linestyle = "-.")
yeniEksen.plot(numpyDizi3, numpyDizi3 + 5, color = "red", linestyle = ":")
yeniEksen.plot(numpyDizi3, numpyDizi3 + 6, color = "red", linestyle = "--")

yeniEksen.plot(numpyDizi3, numpyDizi3 + 7, color = "yellow", linestyle = "--",linewidth = 5.0, marker = "o",markersize = 4, markerfacecolor = "red")
```
